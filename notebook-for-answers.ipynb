```python
#import ....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#VAT algorithm provided for your use
def VAT(R):
    """

    VAT algorithm adapted from matlab version:
    http://www.ece.mtu.edu/~thavens/code/VAT.m

    Args:
        R (n*n double): Dissimila|rity data input
        R (n*D double): vector input (R is converted to sq. Euclidean distance)
    Returns:
        RV (n*n double): VAT-reordered dissimilarity data
        C (n int): Connection indexes of MST in [0,n)
        I (n int): Reordered indexes of R, the input data in [0,n)
    """
        
    R = np.array(R)
    N, M = R.shape
    if N != M:
        R = squareform(pdist(R))
        
    J = list(range(0, N))
    
    y = np.max(R, axis=0)
    i = np.argmax(R, axis=0)
    j = np.argmax(y)
    y = np.max(y)


    I = i[j]
    del J[I]

    y = np.min(R[I,J], axis=0)
    j = np.argmin(R[I,J], axis=0)
    
    I = [I, J[j]]
    J = [e for e in J if e != J[j]]
    
    C = [1,1]
    for r in range(2, N-1):   
        y = np.min(R[I,:][:,J], axis=0)
        i = np.argmin(R[I,:][:,J], axis=0)
        j = np.argmin(y)        
        y = np.min(y)      
        I.extend([J[j]])
        J = [e for e in J if e != J[j]]
        C.extend([i[j]])
    
    y = np.min(R[I,:][:,J], axis=0)
    i = np.argmin(R[I,:][:,J], axis=0)
    
    I.extend(J)
    C.extend(i)
    
    RI = list(range(N))
    for idx, val in enumerate(I):
        RI[val] = idx

    RV = R[I,:][:,I]
    
    return RV.tolist(), C, I

```

```python
#Q2.1

import json 

with open('roads.json') as f:
    Data = json.load(f)

#access json for road_segment numbers and csv file's data section 
road_list = Data["data"]
road_seg = traffic["road_segment"]

#constants
ROAD_SEGMENT_COL = 9 
STR_TYPE_COL = 14 

#dictionary of strtype with their road segments
str_type = dict()
str_type_list =[]

for i in range(len(road_list)):  
    if road_list[i][STR_TYPE_COL] not in str_type: 
        str_type[road_list[i][ROAD_SEGMENT_COL]] = road_list[i][STR_TYPE_COL] 

#map road_segment in traffic to their street types        
for i in range(len(road_seg)): 
    if str(road_seg.iloc[i]) in str_type.keys(): 
        str_type_list.append(str_type[str(road_seg.iloc[i])])
        
#add StrType attribute to traffic
traffic["StrType"] = str_type_list

print(f"***\nQ2.1\nThe first three rows of traffic DataFrame with the attribute StrType are:\n{traffic.head(3)}\n***")
traffic.to_csv('check2.csv')
```

### Q3.3b

```python
#Q3.3b


%matplotlib inline
import calendar
from numpy import arange
from collections import defaultdict as dd

#do kmeans clustering on k=3 

data = special_traffic[[ column for column in (special_traffic.columns) if column not in 'idx StrType']]
kmeans = KMeans(n_clusters=3, max_iter = 1000).fit(data)


#show size of each cluster with barplot


# dictionary to count number of data points in cluster
Cluster_centres = ['Cluster1', 'Cluster2', 'Cluster3']
cluster_centers_val = dd(int)
for i in range(len(kmeans.labels_)): 
    if kmeans.labels_[i] == int(0) : 
        cluster_centers_val['Cluster 1'] +=1
    elif kmeans.labels_[i] == int(1) : 
        cluster_centers_val['Cluster 2'] +=1
    elif kmeans.labels_[i] == int(2) : 
        cluster_centers_val['Cluster 3'] +=1
    
# bar plot
plt.bar(arange(len(cluster_centers_val.values())),cluster_centers_val.values())
plt.xticks(arange(len(Cluster_centres)),Cluster_centres)
plt.title('Clusters vs. size')
plt.xlabel('Clusters',fontSize = 20 )
plt.ylabel('Size', fontSize = 20)
plt.show()
```

### Q3.3c 

A scatter plot can be used to evaluate the K means clustering method effectiveness in how well it separates data by Street type. EVen if we use k = 3, a scatter plot can be expected to show two clusters that are very close together and one that is far away assuming that the K means is effective in separating data by street type. 




```python
from sklearn.cluster import KMeans

#read data 
special_traffic = pd.read_csv('special_traffic.csv').sample(n=1000) 
data = special_traffic[[ column for column in (special_traffic.columns) if column not in 'idx StrType']]

###normalize data###
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
specialt_scaled = min_max_scaler.fit_transform(data)
specialt_normalized = pd.DataFrame(specialt_scaled)

# perform principal component analysis
sklearn_pca = sklearnPCA(n_components=2)   #we want just the first two PCs
reduced_data = sklearn_pca.fit_transform(specialt_normalized)

#find kmeans
kmeans = KMeans(init='k-means++', n_clusters=3).fit(reduced_data)


# plot scatter plot
plt.scatter(reduced_data[:,0], reduced_data[:,1], c=kmeans.labels_.astype(float))
purple_patch = mpatches.Patch(color='purple', label='Cluster 1')
yellow_patch = mpatches.Patch(color='yellow', label='Cluster 2')
green_patch = mpatches.Patch(color='turquoise', label='Cluster 3')
plt.legend(handles=[purple_patch, yellow_patch, green_patch], loc='upper left',fontsize=8)
plt.xlabel('1st Principal Component', fontsize=28)
plt.ylabel('2nd Principal Component', fontsize=28)
plt.show()
```

Using the k means algorithm with k=3, it shows 3 different clusters and does not seem to signify any separation according to the two street types. Thus, the K means model does not group data well based on attribute StrType.
